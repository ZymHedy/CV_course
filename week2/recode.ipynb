```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lenna.jpg',1)
```

```python
def my_show(img):
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
my_show(img)
```

## Gaussian Kernel

```python
g_img = cv2.GaussianBlur(img,(11,11),20)#imgkernelsize
plt.subplot(121)
my_show(img)
plt.subplot(122)
my_show(g_img)
```

```python
kernel_1d = cv2.getGaussianKernel(11,2)#
```

```python
g1_img = cv2.sepFilter2D(img,-1,kernel_1d,kernel_1d)#xy
```

```python
my_show(g1_img)
```

## laplacian

```python
kernel_l = np.array([[0,1,0],[1,-4,1],[0,1,0]])#
```

```python
lap_img = cv2.filter2D(img,-1,kernel_l)#-1img
```

```python
my_show(lap_img)#
```

```python
kernel_strong = np.array([[1,1,1],[1,-8,1],[1,1,1]])#
```

```python
lapStrong_img = cv2.filter2D(img,-1,kernel_strong)#
my_show(lapStrong_img)
```
